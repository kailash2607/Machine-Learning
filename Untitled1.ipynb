```python
# Program developed by : Kailash Prabhu S
# Register Number      : 212224240068

# ==============================================================
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from google.colab import files
import io

# ==============================================================
# Upload CSV manually (handles file name dynamically)
print("ðŸ“‚ Please upload FuelConsumption.csv")
uploaded = files.upload()

filename = list(uploaded.keys())[0]   # take the first uploaded file
print("ðŸ“‚ File uploaded as:", filename)

df = pd.read_csv(io.BytesIO(uploaded[filename]))
print("âœ… File loaded successfully!")
print("First 5 rows of dataset:")
print(df.head())

# ==============================================================
# Helper function to save and show plot
def save_and_show_plot(fig, filename):
    fig.savefig(filename)
    plt.show()

# ==============================================================
# Q1. Scatter plot Cylinder vs Co2Emission (green color)
fig = plt.figure()
plt.scatter(df['CYLINDERS'], df['CO2EMISSIONS'], color='green')
plt.xlabel("Cylinder")
plt.ylabel("CO2 Emission")
plt.title("Cylinder vs CO2 Emission")
save_and_show_plot(fig, "Q1_Cylinder_vs_CO2.png")

# ==============================================================
# Q2. Compare Cylinder vs CO2Emission and EngineSize vs CO2Emission
fig = plt.figure()
plt.scatter(df['CYLINDERS'], df['CO2EMISSIONS'], color='blue', label="Cylinder vs CO2")
plt.scatter(df['ENGINESIZE'], df['CO2EMISSIONS'], color='red', label="EngineSize vs CO2")
plt.xlabel("Cylinder / EngineSize")
plt.ylabel("CO2 Emission")
plt.legend()
plt.title("Cylinder & EngineSize vs CO2 Emission")
save_and_show_plot(fig, "Q2_Cylinder_Engine_vs_CO2.png")

# ==============================================================
# Q3. Compare Cylinder, EngineSize, FuelConsumption_comb vs CO2Emission
fig = plt.figure()
plt.scatter(df['CYLINDERS'], df['CO2EMISSIONS'], color='blue', label="Cylinder vs CO2")
plt.scatter(df['ENGINESIZE'], df['CO2EMISSIONS'], color='red', label="EngineSize vs CO2")
plt.scatter(df['FUELCONSUMPTION_COMB'], df['CO2EMISSIONS'], color='green', label="FuelConsumption vs CO2")
plt.xlabel("Cylinder / EngineSize / FuelConsumption")
plt.ylabel("CO2 Emission")
plt.legend()
plt.title("Multiple Factors vs CO2 Emission")
save_and_show_plot(fig, "Q3_Multiple_vs_CO2.png")

# ==============================================================
# Q4. Train model with Cylinder as independent variable
X_cyl = df[['CYLINDERS']]
y = df['CO2EMISSIONS']

X_train, X_test, y_train, y_test = train_test_split(X_cyl, y, test_size=0.2, random_state=42)
model_cyl = LinearRegression()
model_cyl.fit(X_train, y_train)

y_pred_cyl = model_cyl.predict(X_test)
print("\nModel (Cylinder -> CO2):")
print("Coefficient:", model_cyl.coef_)
print("Intercept:", model_cyl.intercept_)
print("R2 Score:", r2_score(y_test, y_pred_cyl))

# Plot regression line
fig = plt.figure()
plt.scatter(X_test, y_test, color='blue', label="Actual")
plt.plot(X_test, y_pred_cyl, color='red', linewidth=2, label="Regression Line")
plt.xlabel("Cylinders")
plt.ylabel("CO2 Emission")
plt.legend()
plt.title("Cylinder vs CO2 (Regression Line)")
save_and_show_plot(fig, "Q4_Cylinder_Regression.png")

# ==============================================================
# Q5. Train model with FuelConsumption_comb as independent variable
X_fuel = df[['FUELCONSUMPTION_COMB']]
y = df['CO2EMISSIONS']

X_train, X_test, y_train, y_test = train_test_split(X_fuel, y, test_size=0.2, random_state=42)
model_fuel = LinearRegression()
model_fuel.fit(X_train, y_train)

y_pred_fuel = model_fuel.predict(X_test)
print("\nModel (FuelConsumption -> CO2):")
print("Coefficient:", model_fuel.coef_)
print("Intercept:", model_fuel.intercept_)
print("R2 Score:", r2_score(y_test, y_pred_fuel))

# Plot regression line
fig = plt.figure()
plt.scatter(X_test, y_test, color='blue', label="Actual")
plt.plot(X_test, y_pred_fuel, color='red', linewidth=2, label="Regression Line")
plt.xlabel("Fuel Consumption (Comb)")
plt.ylabel("CO2 Emission")
plt.legend()
plt.title("FuelConsumption vs CO2 (Regression Line)")
save_and_show_plot(fig, "Q5_FuelConsumption_Regression.png")

# ==============================================================
# Q6. Train model on different train-test ratios
ratios = [0.2, 0.3, 0.4, 0.5]
print("\nAccuracy for different Train-Test splits:")
for r in ratios:
    X_train, X_test, y_train, y_test = train_test_split(X_fuel, y, test_size=r, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = r2_score(y_test, y_pred)
    print(f"Train-Test Split {1-r:.0%}-{r:.0%} -> Accuracy (R2): {acc:.4f}")
```
